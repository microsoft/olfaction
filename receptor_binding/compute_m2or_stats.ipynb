```python
## Get 'InChi Key' for each 'smiles' in m2or_pairwise_no_mixtures

from rdkit import Chem
def get_inchi_key_rdkit(smiles):
    # Generate an RDKit molecule from the SMILES string
    mol = Chem.MolFromSmiles(smiles)

    if mol:
        # Generate the InChI Key
        inchi_key = Chem.MolToInchiKey(mol)
        return inchi_key

    return None

# Example usage
smiles_lst = m2or_pairwise_no_mixtures['smiles'].tolist()

inchi_key_lst = [get_inchi_key_rdkit(smiles) for smiles in smiles_lst]
len(inchi_key_lst)
```

```python
m2or_pairwise_no_mixtures['InChi Key'] = inchi_key_lst
```
