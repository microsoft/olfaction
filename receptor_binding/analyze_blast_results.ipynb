```python
import pandas as pd


# Read the blast results file
blast_results = pd.read_csv('M2OR/blast_results.txt', sep='\t', header=None, names=['query', 'subject'])

# Extract Uniprot ID and Gene ID from the subject sequence IDs
blast_results['uniprot_id'] = blast_results['subject'].str.split('|').str[1]
blast_results['gene_id'] = blast_results['subject'].str.split('|').str[2]

# Display the resulting dataframe
print(blast_results)
```

```python
blast_results.columns
```

```python
blast_results['uniprot_id'].unique().shape
```

```python
blast_results['sequence'] = blast_results['query']
```

```python
### infil M2OR_2023_04_28_full_seq_mut_canonSMILES.csv with uniprot ids
m2or = pd.read_csv("M2OR/M2OR_2023_04_28_full_seq_mut_canonSMILES.csv")
m2or['Uniprot ID'].unique().shape
```

```python
m2or
```

```python
m2or_nan = m2or[m2or['Uniprot ID'].isna()]
m2or_nonan = m2or[~m2or['Uniprot ID'].isna()]

m2or_nan
```

```python
for row in blast_results.itertuples():
    m2or_nan.loc[m2or['Sequence'] == row.sequence, 'Uniprot ID'] = row.uniprot_id

m2or
```

```python
m2or_nan['Uniprot ID'].isna().sum()
```

```python
## Merge m2or_nonan and m2or_nan to get the full dataframe with uniprot ids
m2or_blast = pd.concat([m2or_nonan, m2or_nan], axis=0)

m2or_blast
```

```python
m2or_blast.to_csv('M2OR/M2OR_2023_04_28_full_seq_mut_canonSMILES_Uniprot_ID.csv', index=False)
```
